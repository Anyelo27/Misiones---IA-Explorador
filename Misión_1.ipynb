# **Taller: Análisis Exploratorio de Datos (EDA) con Dataset de Películas**

## **Importando librerías**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Carga de los datos**

```python
dataframe = pd.read_csv('movie_metadata.csv')
```

## **Revisión General de los Datos**

```python
print('Las primeras 5 filas del DataFrame son:')
dataframe.head()
```

```python
print('La información del Dataframe inicialmente es:')
dataframe.info()
print('\n')
print('La descripción del Dataframe inicialmente es:')
dataframe.describe()
```

## **Tratamiento de datos Duplicados**
###(Búsqueda y Eliminación de Duplicados)


```python
# Búsqueda de Duplicados
duplicados = dataframe.duplicated().sum()
print('Hay', duplicados, 'registros duplicados, estos son:')
print('\n')
dataframe[dataframe.duplicated()]
```

```python
# Eliminación de Duplicados
dataframe_cleaned = dataframe.drop_duplicates()
print('La información del Dataframe sin duplicados es:')
print('\n')
dataframe_cleaned.info()
print('\n')
print('Fueron eliminados', duplicados, 'registros duplicados')
print('Quedan', {dataframe_cleaned.shape[0]} ,'filas restantes')
```

## **Análisis de entradas sin información**
### (Identificación de Valores Nulos)

```python
print('Los valores nulos del Dataframe son:')
print('\n')
dataframe_cleaned.isnull().sum()
```

## **Visualización de Variables**

```python
plt.hist(dataframe_cleaned['imdb_score'], bins=20, edgecolor='black')
plt.title('Número de películas por IMBD')
plt.xlabel('IMBD Score')
plt.ylabel('Cantidad de películas')
plt.grid(True)
plt.show()

dataframe_cleaned['genres'].value_counts().head(10).plot(kind='bar')
plt.title('Top 20 géneros de películas')
plt.xlabel('Géneros')
plt.ylabel('Cantidad de películas')
plt.grid(True)
plt.show()
```

## **Análisis de Correlaciones**

```python
# Corrigiendo el tipo de variable Budget
dataframe_cleaned['budget'] = dataframe_cleaned['budget'].str.replace('$', '')

correlacion = dataframe_cleaned[['budget', 'gross', 'imdb_score']].corr()
print('La correlación entre Budget, Gross e IMDB Score es:')
print('\n')
print(correlacion)
print('\n')
sns.heatmap(correlacion, annot=True)
plt.title('Correlación entre Budget, Gross e IMDB Score')
plt.show()
```

## **Gráficos de Dispersión**

```python
dataframe_reducido = dataframe_cleaned.head(50)
plt.figure(figsize=(15, 5))
plt.scatter(dataframe_reducido['gross'], dataframe_reducido['budget'])
plt.title('Relación entre Gross y Budget de películas')
plt.xlabel('Gross')
plt.ylabel('Budget')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.scatter(dataframe_reducido['imdb_score'], dataframe_reducido['budget'])
plt.title('Relación entre Budget e IMBD Score de películas')
plt.xlabel('IMBD Score')
plt.ylabel('Budget')
plt.show()
```

## **Tratamiento de entradas sin información**

```python
dataframe_cleaned2 = dataframe_cleaned.dropna()
print('La información del Dataframe sin valores nulos es:')
print('\n')
dataframe_cleaned2.info()
```

## **Visualización de Variables del DataSet sin Valores Nulos**

```python
plt.hist(dataframe_cleaned2['imdb_score'], bins=20, edgecolor='black')
plt.title('Número de películas por IMBD')
plt.xlabel('IMBD Score')
plt.ylabel('Cantidad de películas')
plt.grid(True)
plt.show()

dataframe_cleaned2['genres'].value_counts().head(10).plot(kind='bar')
plt.title('Top 20 géneros de películas')
plt.xlabel('Géneros')
plt.ylabel('Cantidad de películas')
plt.grid(True)
plt.show()
```

## **Gráficos de Dispersión del Dataset sin Valores Nulos**
