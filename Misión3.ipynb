3. Tokenizaci√≥n:

Emplea nltk para dividir el texto en tokens:

```python
import nltk
from nltk.tokenize import word_tokenize

# Descargar recursos necesarios
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')


# Tokenizar una rese√±a de ejemplo
example_review = df['reviews.text'][0]
tokens = word_tokenize(example_review)
print(tokens)
```

4. Normalizaci√≥n:

Convierte el texto a min√∫sculas y elimina caracteres especiales:

```python
#Texto a min√∫sculas
texto_normalizado = df['reviews.text'].str.lower()

#Eliminar caracteres especiales
texto_normalizado = texto_normalizado.apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
texto_normalizado = texto_normalizado.apply(lambda x: unicodedata.normalize('NFKD', x))
texto_normalizado = texto_normalizado.apply(lambda x: x.encode('ASCII', 'ignore').decode('utf-8'))


print("Texto normalizado:", texto_normalizado)
```

5. Eliminaci√≥n de Stopwords:

Utiliza **nltk** para eliminar palabras comunes que no aportan valor sem√°ntico:

```python
# Descargar lista de stopwords
tokens = word_tokenize(texto_normalizado[0])
stop_words = set(stopwords.words('spanish'))
tokens_filtrados = [palabra for palabra in tokens if palabra.lower() not in stop_words]
texto_filtrado = ' '.join(tokens_filtrados)
print("Texto sin stopwords:", texto_filtrado)
```

6. Lematizaci√≥n:

Aplica WordNetLemmatizer de nltk para reducir las palabras a su forma base:

```python
!python -m spacy download es_core_news_sm
nlp = spacy.load('es_core_news_sm')
texto_lematizacion = nlp(texto_filtrado)
tokens_lematizados = [(token.text, token.lemma_) for token in texto_lematizacion ]

print("Texto original:", texto_filtrado)
print("Lematizaci√≥n:")
for palabra, lema in tokens_lematizados:
    print(f"{palabra} -> {lema}")
```

7. Stemming:

Alternativamente, utiliza **PorterStemmer** para reducir las palabras a su ra√≠z:

```python
stemmer = SnowballStemmer("spanish")
texto = texto_filtrado
tokens = word_tokenize(texto)
stems = [(palabra, stemmer.stem(palabra)) for palabra in tokens]
print("Texto original:", texto)
print("Stemming:")
for palabra, raiz in stems:
    print(f"{palabra} -> {raiz}")
```
